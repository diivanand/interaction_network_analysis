```python
#import necessary modulesimport numpy as npfrom datetime import datetime, timedeltafrom dateutil.parser import parsefrom pandas import *import bisectimport sysimport ast
```

```python
#import data from the csv file that is the output of daily_interaction_pairs.piginput = read_csv(filepath_or_buffer="inter_pairs_art_daily_cnts.csv",sep=';',names=['DATE','EDGE','CREATIONTYPE','RESPONSETYPE','CNT'])input
```

```python
#convert the DATA and EDGE columns from string datatypes into datetime and tuple object datatypes respectivelyinput.DATE = input.DATE.apply(parse)input.EDGE = input.EDGE.apply(ast.literal_eval)input
```

```python
ctype = ''rtype = ''window = 28threshold = 1
```

```python
if len(ctype) > 0 and len(rtype) > 0:    data = input[input['CREATIONTYPE'] == ctype]    data = data[data['RESPONSETYPE'] == rtype]    data = data.pivot(index='DATE', columns='EDGE', values='CNT')else:    ctype = 'ALL'    rtype = 'ALL'    data = pivot_table(data=input,values='CNT',rows='DATE',cols='EDGE',aggfunc=sum)data
```

```python
countsTS = data.sum(axis=1)countsTS = countsTS.select(lambda dt: dt.weekday() < 5)countsTS.plot()title(ctype + "-" + rtype + " Interation Pair Counts")ylabel("Daily Interaction Pair Counts")
```

```python
#Apply rolling sumrsData = rolling_sum(data,window,min_periods=1)q50 = rsData.quantile(q=.5,axis=1)q70 = rsData.quantile(q=.7,axis=1)q90 = rsData.quantile(q=.9,axis=1)df = DataFrame({"q50":q50,"q70":q70,"q90":q90})df.plot(ylim=(0,4),yticks=[0,1,2,3,4,5])title(ctype + "-" + rtype + " Sustained Interaction Counts Distribution")ylabel("Number of interactions within window period")
```
